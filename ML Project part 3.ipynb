```python
import numpy as np 
import pandas as pd

def get_labelled_data(file):
    f = open(file,'r',encoding = 'utf-8')
    lines = f.readlines()
    data = []
    
    start = 0
    for i in range(len(lines)):
        if lines[i] == '\n':
            data.append(lines[start:i])
            start = i+1
        lines[i] = lines[i].replace('\n','')
        lines[i] = tuple(lines[i].split(' '))
        
    output = {}    
        
    for i in range(len(data)):
        newdata = data[i]
        for j in range(len(newdata)):
            tup = newdata[j]
            word = tup[0]
            tag = tup[1]
            if word not in output:
                output[word] = {tag:1}
            else:
                if tag not in output[word]:
                    output[word][tag] = 1
                else:
                    output[word][tag] += 1
        
    return output 
    
d = get_labelled_data('ML Project/EN/train')
print (d)
```
